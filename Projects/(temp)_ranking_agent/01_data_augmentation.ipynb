```python
import pandas as pd
import numpy as np
import random
import datetime
import os

FINAL_BRAND_NAME = "LANEIGE(ラネージュ)"

# ----------------------------------------
# 날짜 관련 함수
# ----------------------------------------

def random_collect_date(base_date_str):
    if "/" in base_date_str:
        fmt = "%Y/%m/%d"
    else:
        fmt = "%Y%m%d"

    base_date = datetime.datetime.strptime(base_date_str, fmt)
    delta_days = random.randint(-5, 5)
    return (base_date + datetime.timedelta(days=delta_days)).strftime("%Y/%m/%d")


def get_friday_of_week(date_str):
    """YYYY/MM/DD → 그 주 금요일(collect_date)"""
    dt = datetime.datetime.strptime(date_str, "%Y/%m/%d").date()
    friday = dt + datetime.timedelta(days=(4 - dt.weekday()))
    return friday.strftime("%Y/%m/%d")


def random_start_end_for_top100():
    """
    종합뷰티랭킹 날짜 규칙:
    - start_date = 목요일
    - end_date = 다음 주 수요일
    - 총 7일간의 수집 기간
    """
    # 최근 6개월 범위에서 날짜 랜덤 선택
    base = datetime.date.today() - datetime.timedelta(days=random.randint(1, 180))

    # base 날짜가 속한 주의 목요일을 찾기
    # weekday(): Mon=0, Tue=1, Wed=2, Thu=3, Fri=4, Sat=5, Sun=6
    offset_to_thu = (3 - base.weekday()) % 7
    start_dt = base + datetime.timedelta(days=offset_to_thu)

    # end_date = start_date + 6일 (목→수)
    end_dt = start_dt + datetime.timedelta(days=6)

    return start_dt.strftime("%Y/%m/%d"), end_dt.strftime("%Y/%m/%d")



def random_start_end_default(collect_date):
    """일반 카테고리: 기존 로직 유지"""
    collect_dt = datetime.datetime.strptime(collect_date, "%Y/%m/%d").date()
    end_dt = collect_dt - datetime.timedelta(days=random.randint(3, 10))
    start_dt = end_dt - datetime.timedelta(days=random.randint(3, 7))
    return start_dt.strftime("%Y/%m/%d"), end_dt.strftime("%Y/%m/%d")


def valid_release_date(start_date, collect_date, base_release_date=None):
    """release_date 규칙 유지 + 실제 기반 분포 적용"""
    start_dt = datetime.datetime.strptime(start_date, "%Y/%m/%d").date()
    collect_dt = datetime.datetime.strptime(collect_date, "%Y/%m/%d").date()
    latest_allowed = min(start_dt, collect_dt)

    # 실제 제품 출시일 기반 사용 시도
    if base_release_date:
        try:
            rd = datetime.datetime.strptime(base_release_date, "%Y/%m/%d").date()
            if rd < latest_allowed:
                return base_release_date
        except:
            pass

    # 규칙 위반 시 랜덤 생성
    release_dt = latest_allowed - datetime.timedelta(days=random.randint(30, 900))
    return release_dt.strftime("%Y/%m/%d")


# ----------------------------------------
# Rank 및 Product 분포 가중치
# ----------------------------------------

def weighted_random_product(product_list):
    """제품 weight 기반 확률 선택"""
    weights = [p.get("weight", 1.0) for p in product_list]
    return random.choices(product_list, weights=weights, k=1)[0]


def weighted_random_rank(rank_limit):
    """상위 rank가 더 자주 등장하도록 Zipf-like 분포 적용"""
    ranks = list(range(1, rank_limit + 1))
    weights = [1 / (r ** 1.2) for r in ranks]
    return random.choices(ranks, weights=weights, k=1)[0]


def get_rank_limit(sheet_name):
    return 100 if "종합뷰티랭킹" in sheet_name else 50


# ----------------------------------------
# 기본 데이터 생성 함수
# ----------------------------------------

def generate_base_rows(sheet, product_pool, n_rows=10):
    rows = []

    for _ in range(n_rows):

        if "종합뷰티랭킹" in sheet:
            start_date, end_date = random_start_end_for_top100()
            collect_date = get_friday_of_week(end_date)
        else:
            # collect_date 생성
            collect_date = random_collect_date("20250101")
            start_date, end_date = random_start_end_default(collect_date)

        product = weighted_random_product(product_pool)
        release_date = valid_release_date(start_date, collect_date, product["release_date"])

        rows.append({
            "collect_date": collect_date,
            "start_date": start_date,
            "end_date": end_date,
            "rank": weighted_random_rank(get_rank_limit(sheet)),
            "brand": FINAL_BRAND_NAME,
            "product_name": product["product_name"],
            "release_date": release_date,
        })

    return pd.DataFrame(rows)

def generate_rows_to_reach_target(sheet, product_pool, target_n, existing_df):
    """기존 데이터 개수에 맞춰 부족한 만큼만 생성하여 총 target_n을 맞춤"""
    
    current_n = len(existing_df)
    if current_n >= target_n:
        return existing_df.head(target_n)

    needed = target_n - current_n
    print(f"→ {sheet}: 기존 {current_n}개, 생성 필요 {needed}개")

    new_rows = []

    for _ in range(needed):
        if "종합뷰티랭킹" in sheet:
            start_date, end_date = random_start_end_for_top100()
            collect_date = get_friday_of_week(end_date)
        else:
            collect_date = random_collect_date("20250101")
            start_date, end_date = random_start_end_default(collect_date)

        chosen = weighted_random_product(product_pool)
        product_name = chosen["product_name"]
        base_release = chosen["release_date"]

        release_date = valid_release_date(start_date, collect_date, base_release)
        rank = weighted_random_rank(get_rank_limit(sheet))

        new_rows.append({
            "collect_date": collect_date,
            "start_date": start_date,
            "end_date": end_date,
            "rank": rank,
            "brand": FINAL_BRAND_NAME,
            "product_name": product_name,
            "release_date": release_date,
        })

    return pd.concat([existing_df, pd.DataFrame(new_rows)], ignore_index=True)


# ----------------------------------------
# 메인 처리 로직
# ----------------------------------------

input_file = "data/COSME_LANEIGE_Ranking_History.xlsx"
output_file = "data/COSME_LANEIGE_Ranking_History_AUGMENTED.xlsx"

xls = pd.ExcelFile(input_file)
augmented_data = {}
TARGET_COUNT = 100  # 시트별 최종 row 개수

for sheet in xls.sheet_names:

    print(f"\n▶ Processing sheet: {sheet}")

    df = pd.read_excel(xls, sheet_name=sheet)

    category_key = None
    for key in CATEGORY_PRODUCT_DB:
        if key in sheet:
            category_key = key
            break

    product_pool = CATEGORY_PRODUCT_DB[category_key]

    if df.empty:
        df = generate_base_rows(sheet, product_pool, n_rows=TARGET_COUNT)

    df = generate_rows_to_reach_target(
        sheet=sheet,
        product_pool=product_pool,
        target_n=TARGET_COUNT,
        existing_df=df
    )

    augmented_data[sheet] = df


with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    for sheet, data in augmented_data.items():
        data.to_excel(writer, sheet_name=sheet, index=False)

print("\n 저장 파일:", output_file)

```
