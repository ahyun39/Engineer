```python
import numpy as np
import matplotlib.pyplot as plt
```

### Step Function

```python
def step(x):
    return np.array(x > 0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step(x)

plt.title("Step Function")
plt.plot(x, y)
plt.show()
```

### Sigmoid Function

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.plot([0, 0],[1.0, 0.0], ':')
plt.title("Sigmoid Function")
plt.show()
```

### Tanh Function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y)
plt.plot([0, 0], [1.0, -1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title("Tanh Function")
plt.show()
```

sigmoid function보다 기울기 소실이 더 적어서 은닉층에서 더 선호.

### ReLU Function

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0, 0], [5.0, 0.0], ':')
plt.title("ReLU Function")
plt.show()
```

0 이상의 입력값의 경우에는 미분값이 항상 1이다.

입력값이 음수면 기울기인 미분값이 0이 되는 문제가 있다.

### Leaky ReLU

```python
a = 0.1

def leaky_relu(x):
    return np.maximum(a*x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.plot([0, 0], [5.0, 0.0], ':')
plt.title('Leaky ReLU Function')
plt.show()
```

### Softmax Function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
```
